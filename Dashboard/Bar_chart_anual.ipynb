# Annual alerts

```python
data = pd.read_csv('./../full_RASFF_DATA.csv', sep=';', 
                   header=0, index_col = 0)
```

```python
data['DATE_CASE'] =pd.to_datetime(data.DATE_CASE)
datag = data.groupby(data.DATE_CASE.dt.year).size()
```
